```
# TODO

# Compute the continuous valued feature list
feat_cont = floats

#Compute the categorical valued feature list
#Remove the class label column(A16)
feat_cat = list(set(col_names) - set(feat_cont))
feat_cat.remove('A16')

print(feat_cont, feat_cat)
```

```
#Group by the credit class attribute(A16)
grouped = credit_df.groupby(credit_df.A16)

#Extract the entries in the positive class
pos_credit_df = grouped.get_group('+').drop(columns=['A16'])

#Extract the entries in the negative class
neg_credit_df = grouped.get_group('-').drop(columns=['A16'])

# Verify the entry counts in the positive and negative class dataframes
# by computing the the '+' and '-' value counts.
display(credit_df.A16.value_counts())
display(pos_credit_df)
display(neg_credit_df)
```

```
from scipy.stats import ttest_ind

pvalues = []

# Use 2 sample ttest to see if there is a significant difference for a feature 
# between the 2 credit classes. 
# Store the (feature name, pvalue) tuple in the pvalues list
for feat in feat_cont:
  pvalues.append((feat,ttest_ind(pos_credit_df[feat], neg_credit_df[feat], nan_policy='omit').pvalue))

#Sort the pvalues list on the pvalue for each feature
pvalues.sort(key=lambda x:x[1])
display(pvalues)
```

```
from scipy.stats import chi2_contingency

chi2_pvalues = []

# Use the chi2_contingency function to see is there is any correlation
# between a feature and the credit class
# Store the (feature name, pvalue) tuple in the chi2_pvalues list
for feat in feat_cat:
  _,pvalue,_,_ = chi2_contingency(pd.crosstab(credit_df[feat], credit_df.A16))
  chi2_pvalues.append((feat, pvalue))

#Sort the chi2_pvalues list on the pvalue for each feature
chi2_pvalues.sort(key=lambda x:x[1])
display(chi2_pvalues)
```

## Part 4 - Bayesian vs Frequentist Statistics

Using a minimum of 2-3 sentences, give an explanation of Bayesian and Frequentist statistics, and then compare and contrast these two approaches to statistical inference.



Frequentist statistics computes probabilities using observable counts/frequencies e.g. there are six numbers on a die, so the probability of getting
any one of those numbers(assuming a fair die) should be 1/6; throw a die N times and observe the count of the number of times each of the numbers appear $n_1$,...,$n_6$, the probability $p_i$ for each number will be $p_i$ = $n_i$/N.

Bayesian statistics tell you how to compute the probability of an event(H), given that another event(E) has occurred by updating the prior(previously known) probability of H to incorporate the fact that E has occurred.

Bayesian and Frequentist statistics are complementary in the sense that the prior probabilities used in Bayesian statistics are normally provided by Frequentist techniques.
